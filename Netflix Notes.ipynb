```python
import pandas as pd
```

```python
file = pd.read_csv('ViewingActivity-sample.csv')
```

```python
file = file.drop(['Profile Name', 'Attributes', 'Supplemental Video Type', 'Device Type', 'Bookmark', 'Latest Bookmark', 'Country'], axis=1)
#Dropping columns that are not important to our data---<file var>.drop([list of desired dropped columns],axis=1) axis = 1 tells pandas to drop that column.
```

```python
file #currently the start time and duration (type) == object meaning they are strings. To analyze the data we need 
######to convert the time to Datetime and Timedelta in Pandas
######Cant work with times as a string
######Step 1: Convert Startime from an Object to Datetime so Pandas can perform calculations
######Step 2: Start time will be in UTC time zone so we have to change it to my local time zone EST
######Step 3: Convert Duration to time Delta so Pandas can perform calculations
```

```python
#Step 1
file['Start Time'] = pd.to_datetime(file['Start Time'], utc=True)
#used pd.to_datetime()
#UTC is for the time zone identification, started with UTC so we can convert to EST which is done with...
#.tz_convert() catch is that tz.convert can only be used with a Datetimeindex so Start Time first has
#to be set as the index usingn set_index() then we will undo using. reset_index()
```

```python
file = file.set_index('Start Time')
```

```python
file.index = file.index.tz_convert('US/Eastern')
```

```python
file =  file.reset_index()
```
